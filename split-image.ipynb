```python
import cv2
from PIL import Image
import os
```

```python
PATH = "E:/inria/AerialImageDataset/train/gt/"
for image in os.listdir("E:/inria/AerialImageDataset/train/gt/"):
    im = cv2.imread(PATH + image)
    tiles = [im[x:x+500,y:y+500] for x in range(0,im.shape[0],500) for y in range(0,im.shape[1],500)]
    k = 0
    for tile in tiles:
        outimg = Image.fromarray(tile, "RGB" )
        index = image.find('.tif')
        newname = image[:index] + '_' + str(k) + image[index:]
        outimg.save("E:/inria/AerialImageDataset/train/seg/" + newname)
        print("Save ", newname)
        k= k + 1
```
